```python
#import the cleaned data file
import pandas as pd
df = pd.read_excel('C:/Users/ashpi/Etsy_Reveiw_Data_2.xlsx')
df.head()
```

```python
# Imports

import matplotlib.pyplot as plt
import seaborn as sns
color = sns.color_palette()
%matplotlib inline
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tls
import plotly.express as px

# Product Scores
fig = px.histogram(df, x="Individual_Star_rating")
fig.update_traces(marker_color="pink",marker_line_color='rgb(50,100,0)',
                  marker_line_width=2)
fig.update_layout(title_text='Individual_Star_rating')
fig.show()
```

```python
from nltk.corpus import stopwords
from wordcloud import WordCloud

# Create stopword list:
stopwords = set(stopwords.words('english'))
stopwords.update(["br", "href"])
textt = " ".join(df.Review.apply(str))
wordcloud = WordCloud(stopwords=stopwords).generate(textt)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.savefig('wordcloud_etsy11.png')
plt.show()
```

```python
# Define positive and negative sentiment where postive is when rating is greater than 3

df['sentiment'] = df['Individual_Star_rating'].apply(lambda rating : +1 if rating > 3 else -1)
positive = df[df['sentiment'] == 1]
negative = df[df['sentiment'] == -1]
```

```python
stopwords.update(["br", "href","book","ordered"]) 
## book and ordered removed because they were included in negative sentiment

# Wordcloud for positive sentiment
pos = " ".join(positive.Review.apply(str))
wordcloud2 = WordCloud(stopwords=stopwords).generate(pos)
plt.imshow(wordcloud2, interpolation='bilinear')
plt.axis("off")
plt.savefig('wordcloud_etsy22.png')
plt.show()
```

```python
# Wordcloud for negative sentiment
neg = " ".join(negative.Review.apply(str))
wordcloud2 = WordCloud(stopwords=stopwords).generate(neg)
plt.imshow(wordcloud2, interpolation='bilinear')
plt.axis("off")
plt.savefig('wordcloud_etsy33.png')
plt.show()
```
